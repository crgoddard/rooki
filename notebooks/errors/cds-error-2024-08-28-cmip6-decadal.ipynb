# CDS error 2024-08-28

We have had an issue raised by a user. They noticed that for the DCPP dataset in the CDS (old CDS, new CDS - both have the issue): the 1976 'base year' for EC-Earth3 monthly sea level pressure can't be retrieved.

This seems to be an issue/inconsistency with the time coordinates in the underlying data for this initialization only, as all base years before and after this can be downloaded. The other two variables, and daily data, also fail.

Their request was as follows:
```
'projections-cmip6-decadal-prototype',
    {
        'temporal_resolution': 'monthly',
        'variable': 'sea_level_pressure',
        'model': 'ec_earth3',
        'base_year': '1976',
        'year': [
            '1976', '1977', '1978',
            '1979', '1980', '1981',
            '1982', '1983', '1984',
            '1985', '1986', '1987',
        ],
        'month': [
            '01', '02', '03',
            '04', '05', '06',
            '07', '08', '09',
            '10', '11', '12',
        ],
        'format': 'zip',
    },
    'download.zip'
```

The error returned was:
Cannot combine along dimension time with mixed types. Found: DatetimeProlepticGregorian, DatetimeGregorian. If importing data directly from a file then setting use_cftime=True may fix this issue.

As an aside, in the new CDS the error returned is less informative, simply: 'The job failed with: RoocsRuntimeError'
